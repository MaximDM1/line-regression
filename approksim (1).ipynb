```python
import numpy as np
from matplotlib import pylab as plt
import scipy
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
def polin(*args):
    y = [f(i) for i in args]
    shp = len(y)
    y = np.array(y)
    b = y.reshape((shp, 1))
    A = [[x**i for i in range(shp)] for x in args]
    A = np.array(A)
    w = scipy.linalg.solve(A, b)
    x = np.arange(1, 15, 0.1)
    f_p = sum([w[i] * x**i for i in range(shp)])
    plt.plot(x, f(x))
    plt.plot(x, f_p)
    plt.show()
```

```python
polin(1, 6, 12)
```

```python
polin(1, 6, 12, 15)
```

```python
polin(1, 3, 6, 12, 15)
```
