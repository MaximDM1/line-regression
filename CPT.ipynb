```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
a = 1
b = 4
uniform_rv = sts.uniform(a, b - a)
x = np.arange(0, 6, 0.01)
y = uniform_rv.pdf(x)
h = uniform_rv.rvs(10000)
plt.plot(x, y)
plt.hist(h, density=True)
```

```python
n = 40
v = [[np.random.choice(h) for _ in range(n)] for _ in range(1000)]
sr = np.array([sum(v[i]) / n for i in range(1000)])
plt.hist(sr, density=True)
```
